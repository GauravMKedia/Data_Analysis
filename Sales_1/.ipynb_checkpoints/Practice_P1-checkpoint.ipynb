```python
# Here We get the last Purchase Date of customer and Recency means how many data before the last date of Purchase has the customer made his lst purchase
df_recency = csv_file.groupby(by='customer_name',as_index=False)['order_date'].max()
df_recency.columns = ['CustomerName', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
# print(recent_date)
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(lambda x: (recent_date-x).days)
df_recency
```

```python
df_recency=df_recency.sort_values(by = 'Recency')
df_recency
```

```python
# Frequency Tell us how many time cutomer has visited and purchased
frequency_df = csv_file.drop_duplicates().groupby(by=['customer_name'], as_index=False)['order_date'].count()
frequency_df.columns = ['CustomerName', 'Frequency']
frequency_df
```

```python
frequency_df=frequency_df.sort_values(by = 'Frequency',ascending=False)
frequency_df
```

```python
csv_file['Total'] = csv_file['sales']
monetary_df = csv_file.groupby(by='customer_name', as_index=False)['Total'].sum()
monetary_df.columns = ['CustomerName', 'Monetary']
monetary_df=monetary_df.sort_values(by = 'Monetary',ascending=False)
monetary_df
```

```python
rf_df = df_recency.merge(frequency_df, on='CustomerName')
rfm_df = rf_df.merge(monetary_df, on='CustomerName').drop(columns='LastPurchaseDate')
rfm_df_monetary=rfm_df.sort_values(by='Monetary',ascending=False)
print(rfm_df_monetary.head())
# rfm_df_frequency=rfm_df.sort_values(by='Frequency',ascending=False)
# print(rfm_df_frequency.head())
# rfm_df_recency=rfm_df.sort_values(by='Recency')
# print(rfm_df_recency.head())
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)

# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100

rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

rfm_df.head()
```
